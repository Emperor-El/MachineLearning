### Impementing Gradient Descent

```python
# Importing libraries
import numpy as np


# Assuming values for the Independent and dependent variable 

x = [1,2,3,4,5]

y = [y_value*2+3 for y_value in x]  

# Converting list to numpy arrays
x = np.array(x)
y = np.array(y)
```

### Gradient Descent



```python

# Initializing valuse for slope and intercept

m_curr = 0
b_curr = 0


# Gradient Descent function

def gradient_descent(x, y):
    global m_curr
    global b_curr
    iteration = 1000
    
    for i in range(iteration):
        n = len(x)
        learning_rate = 0.02

        y_predict = m_curr*x + b_curr
        cost_func = 1/n * sum([val **2 for val in (y-y_predict)])

        md = -2/n * sum(x*(y-y_predict))
        bd = -2/n * sum(y-y_predict)

        m_curr = m_curr - learning_rate*md
        b_curr = b_curr - learning_rate*bd

        print(f'slope: {m_curr}  intercept: {b_curr}  cost: {cost_func} {i+1}')





gradient_descent(x,y)
```
